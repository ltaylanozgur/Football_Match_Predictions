# Create statistics for seasons 2013-2022

```python
data['team'].unique()
```

```python
def stats(t):
    
    win = 100*len(data[(data['team'].isin([t])) & (data['result']=='W')])/len(data[(data['team'].isin([t]))])
    draw = 100*len(data[(data['team'].isin([t])) & (data['result']=='D')])/len(data[(data['team'].isin([t]))])
    lose = 100*len(data[(data['team'].isin([t])) & (data['result']=='L')])/len(data[(data['team'].isin([t]))])
    goal = data.loc[data['team'] == t, 'gf'].mean()
    ast = data.loc[data['team'] == t, 'ast'].mean()
    error = data.loc[data['team'] == t, 'err'].mean()
    red_card = data.loc[data['team'] == t, 'crdr'].mean()
    yellow_car = data.loc[data['team'] == t, 'crdy'].mean()
    s_def = data.loc[data['team'] == t, 's_def'].mean()
    saves = data.loc[data['team'] == t, 'saves'].mean()
    clean_sheet = data.loc[data['team'] == t, 'cs'].mean()
    
    
    return (round(win,2), round(draw,2), round(lose,2), round(goal,2), round(ast,2), round(error,2), round(red_card,2), round(yellow_car,2), round(s_def,2), round(saves,2), round(clean_sheet,2),  len(data[(data['team'].isin([t]))]), t)
```

```python
epl_standings_copy = epl_standings.copy()
```

```python
epl_standings_copy.head()
```

```python
def champion_searcher(df):

    df['number_of_championships'] = 0 #creating an empty column
        
    for col in epl_standings_copy.columns:
        
        for i in df.index:
            if df[col][i] == 1:
                df['number_of_championships'][i] = df['number_of_championships'][i] + 1
            else:
                df['number_of_championships'][i] = df['number_of_championships'][i]
  
    return df
```

```python
epl_standings_new = champion_searcher(epl_standings_copy)
```

```python
epl_standings_new
```

```python
def relegation_searcher(df):

    df['number_of_relegations'] = 0 #creating an empty column
        
    for col in epl_standings_copy.columns:
        
        for i in df.index:
            if ((df[col][i] == 18) | (df[col][i] == 19) | (df[col][i] == 20)):
                df['number_of_relegations'][i] = df['number_of_relegations'][i] + 1
            else:
                df['number_of_relegations'][i] = df['number_of_relegations'][i]
  
    return df
```

```python
epl_standings_new = relegation_searcher(epl_standings_new)
```
